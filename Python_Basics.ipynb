3. Compare and contrast mutable and immutable objects in Python with examples.

```python
# Examples
'''
Mutable example
list1 = [1, 2, 3]
list1[0] = 10      # Allowed

Immutable example
tuple1 = (1, 2, 3)
tuple1[0] = 10   # Error: not allowed
'''
```

4. Discuss the different types of operators in Python and provide examples of how they are used.
- Python provides several types of operators to perform operations on variables and values.
   - Arithmetic operators (+, -, *, /) perform mathematical operations
   - Comparison operators (==, >, <) compare values.
   - Logical operators (and, or, not) combine conditions.
   - Assignment operators (=, +=) assign or update values.
   - Bitwise operators (&, |) work on binary data.
   - Membership operators (in, not in) check for element presence.
   - Identity operators (is, is not) compare object identities.

5. Explain the concept of type casting in Python with examples.
- Type casting is the process of converting a variable from one data type to another. Python supports implicit casting, where conversion happens automatically, and explicit casting, where you convert data manually using functions like int(), float(), or str().

```python
#Examples:
'''
# Implicit casting
x = 5
y = 2.5
z = x + y      # int automatically converted to float → 7.5

# Explicit casting
a = int(3.8)   # float to int → 3
b = str(10)    # int to string → "10"
c = float(5)   # int to float → 5.0

'''
```

7. Describe the different types of loops in Python and their use cases with examples.
- Python uses loops to execute a block of code multiple times. The for loop iterates over a sequence like a list, tuple, or range, e.g., for i in range(5): print(i) prints 0–4. The while loop runs as long as a condition is True, e.g., x = 0; while x < 5: print(x); x += 1. Loops can be controlled using break to exit early, continue to skip the current iteration, and else after loops runs if the loop completes normally. Loops are widely used in tasks like processing lists, reading files, or automating repetitive operations.
