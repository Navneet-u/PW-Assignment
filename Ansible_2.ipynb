```python
# Apache Web Server Playbook
'''
- name: Setup Apache Web Server
  hosts: localhost
  become: yes

  tasks:

    - name: Install Apache
      apt:
        name: apache2
        state: present
        update_cache: yes

    - name: Start Apache Service
      service:
        name: apache2
        state: started
        enabled: yes

    - name: Create Sample Web Page
      copy:
        content: "<h1>Apache Server Configured Using Ansible</h1>"
        dest: /var/www/html/index.html
        '''
```

```python
''' ansible-playbook apache.yml '''
```
