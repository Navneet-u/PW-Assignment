#Introduction to Kubernetes

- Kubernetes Kubernetes is a platform used to deploy and manage containerized applications. It helps in running applications efficiently using containers.

1. Kubernetes Setup using Minikube

- Minikube is used to create a local Kubernetes cluster for learning and practice.

```python
# Start Minikube cluster

minikube start
```

```python
# Check cluster information

kubectl cluster-info
```

2. Application Deployment

- A simple nginx application can be deployed using Kubernetes deployment commands.

```python
# Create deployment

kubectl create deployment myapp --image=nginx
```

```python
# Check deployment

kubectl get deployments
```

3. Resource Management

Kubernetes resources like Pods and Services can be managed using kubectl commands.

```python
# Check pods

kubectl get pods
```

```python
# Expose deployment as service

kubectl expose deployment myapp --type=NodePort --port=80
```

```python
# Check services

kubectl get services
```

4. Helm Charts

- Helm is a package manager for Kubernetes used to install applications easily.

```python
# Install nginx using Helm

helm install my-nginx bitnami/nginx
```

```python
# Check Helm releases

helm list
```

##Result

The Kubernetes cluster was created successfully using Minikube.
The nginx application was deployed and managed using Pods, Services, and Deployments.
Helm was also used to install applications on Kubernetes.
