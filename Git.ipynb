```python
# 4.

      #1. Put master.txt on master branch, stage and commit
      '''
      mkdir project4
      cd project4
      echo "This is master file" > master.txt
      git init
      git add master.txt
      git commit -m "Add master.txt in master branch"
      '''
      #2. Create 3 branches: public 1, public 2 and private
      '''
      git branch public1
      git branch public2
      git branch private
      '''
      #3. Put public1.txt on public 1 branch, stage and commit
      '''
      git checkout public1
      echo "This is public1 file" > public1.txt
      git add public1.txt
      git commit -m "Add public1.txt in public1 branch"
      '''
      #4. Merge public 1 on master branch
      '''
      git checkout master
      git merge public1 -m "Merge public1 into master"
      '''
      #5. Merge public2 on master branch
      '''
      git checkout public2
      echo "This is public2 file" > public2.txt
      git add public2.txt
      git commit -m "Add public2.txt in public2 branch"
      git checkout master
      git merge public2 -m "Merge public2 into master"
      '''
      #6. Edit master.txt on private branch, stage and commit
      '''
      git checkout private
      echo "Updated master file in private branch" > master.txt
      git add master.txt
      git commit -m "Update master.txt in private branch"
      '''
      #7. Update public1 and public2 with new master code in private
      '''
      git checkout public1
      git merge master -m "Merge master into public1"
      git checkout public2
      git merge master -m "Merge master into public2"
      '''
      #8. Also update new master code on master
      '''
      git checkout master
      git merge private -m "Update master with private changes"
      '''
      #9. Finally update all the code on the private branch
      '''
      git checkout private
      git merge master -m "Bring all latest master changes into private"
      '''
```

```python
# 5.
     #1. Create a Git Flow workflow architecture on Git
     '''
     mkdir project5
     cd project5
     git init

     echo "Initial master code" > master.txt
     git add master.txt
     git commit -m "Initial commit on master branch"
     git branch develop
     '''
     #2. Create all the required branches
     '''
     git checkout develop
     git checkout -b feature-login
     '''
     #3. starting from e feature branch, push the branch to the master, following the architecture
     '''
     echo "Login feature code" > login.txt
     git add login.txt
     git commit -m "Add login feature"
     git checkout develop
     git merge feature-login -m "Merge feature-login into develop"
     git checkout -b release-1.0
     echo "Release preparation" > release.txt
     git add release.txt
     git checkout master
     git merge release-1.0 -m "Release 1.0 merged into master"
     git checkout develop
     git merge release-1.0 -m "Merge release-1.0 back into develop"
     '''
     #4. Push an urgent.txt on master using hotfix
     '''
     git checkout master
     git checkout -b hotfix-urgent
     echo "Urgent fix applied" > urgent.txt
     git add urgent.txt
     git commit -m "Apply urgent hotfix"

     git checkout master
     git merge hotfix-urgent -m "Merge urgent hotfix into master"

     git checkout develop
     git merge hotfix-urgent -m "Merge urgent hotfix into develop"

```
