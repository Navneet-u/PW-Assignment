2. Explain the difference between mutable and immutable data types with examples.
- Mutable data types can be changed after creation, while immutable data types cannot be changed once they are created.
### Mutable Data Types

Examples:

- List
- Dictionary
- Set

Example:

```python
my_list = [1, 2, 3]
my_list.append(4)

print(my_list)
```

##Immutable Data Types

Examples:

- String
- Tuple

Example:

```python
name = "Python"

# Strings cannot be changed directly
# name[0] = "J"   ❌ This gives an error

print(name)
```

3. What are the main differences between lists and tuples in Python?
- Lists and tuples are both used to store multiple items in Python, but they have some important differences.

| List | Tuple |
|---|---|
| Mutable (can be changed) | Immutable (cannot be changed) |
| Uses square brackets `[]` | Uses parentheses `()` |
| Slower | Faster |
| More memory usage | Less memory usage |

4. Describe how dictionaries store data?
- Dictionaries store data in the form of key-value pairs.
Each key is unique and is used to access its corresponding value.




```python
student = {
    "name": "Jonney",
    "age": 20,
    "course": "Python"
}

print(student)
```

5. Why might you use a set instead of a list in Python?

- Remove duplicate values automatically
- Perform faster searching and membership testing

Unlike lists, sets only store unique elements.

| String            | List                       |
| ----------------- | -------------------------- |

7. How do tuples ensure data integrity in Python?
- Tuples ensure data integrity because they are immutable, which means their values cannot be changed after creation. This prevents accidental modification of data.

Example.

```python
data = (1, 2, 3)

# Tuples cannot be modified
# data[0] = 10   ❌ This will give an error

print(data)
```

8. What is a hash table, and how does it relate to dictionaries in Python?
- A hash table is a data structure that stores data using key-value pairs and allows fast data retrieval using a hashing function.

In Python, dictionaries are implemented using hash tables.
This helps dictionaries access values quickly using keys.

11. What advantages do dictionaries offer over lists for certain tasks?
- Strings are immutable in Python, which means they cannot be changed after creation. This improves security, performance, and memory efficiency because Python can manage memory more efficiently and prevent accidental changes to data.

12. Describe a scenario where using a tuple would be preferable over a list.
- A tuple is preferable over a list when the data should not be changed after creation. For example, storing fixed values like days of the week, coordinates, or student roll numbers. Since tuples are immutable, they help protect data from accidental modification.

13. How do sets handle duplicate values in Python?
- Sets automatically remove duplicate values in Python. Each element in a set must be unique, so repeated values are stored only once.

14.  How does the “in” keyword work differently for lists and dictionaries?
- In a list, the in keyword checks whether a value exists in the list.
In a dictionary, the in keyword checks whether a key exists in the dictionary, not the value.

15. Can you modify the elements of a tuple? Explain why or why not?
- No, the elements of a tuple cannot be modified because tuples are immutable in Python. Once a tuple is created, its values cannot be changed, added, or removed.

16. What is a nested dictionary, and give an example of its use case?
- A nested dictionary is a dictionary that contains another dictionary inside it. It is used to store complex and organized data, such as student records, employee details, or database-like information.

17.  Describe the time complexity of accessing elements in a dictionary?
- The average time complexity of accessing elements in a dictionary is:

O(1)

This means dictionary elements can usually be accessed very quickly because dictionaries use hash tables internally.

18. In what situations are lists preferred over dictionaries?
- Lists are preferred over dictionaries when:

- Order of elements is important
- Data needs to be accessed using indexes
- Duplicate values are allowed
- Storing a simple collection of items is required

19. Why are dictionaries considered unordered, and how does that affect data retrieval?
- Dictionaries are considered unordered because their data is stored using hash tables rather than fixed positions. Data retrieval is done using keys instead of indexes, which allows fast access to values.

20. Explain the difference between a list and a dictionary in terms of data retrieval.
- In a list, data is retrieved using an index number, while in a dictionary, data is retrieved using a key. Lists are useful for ordered data, whereas dictionaries are useful for fast lookup using keys.
