1. Introduction to Ansible
- Red Hat Ansible is an automation and configuration management tool used to:


Configure servers


Install software


Automate tasks


Manage systems easily


Ansible is agentless, which means no software is required on client machines.

2. Features of Ansible


- Simple to use


- Uses YAML language


- Agentless architecture


- Automates repetitive tasks


- Useful in DevOps and Cloud Computing



3. Install Ansible
- Install on Ubuntu/Linux


```python
#Step 1: Update system packages
'''
sudo apt update
'''
#Step 2: Install Ansible
'''
sudo apt install ansible -y
'''

#Step 3: Verify installation
'''
ansible --version
'''
```

4. Configure Ansible

Inventory file contains system information.

```python
#5. Test Ansible Connection

#Run this command:
'''
ansible localhost -m ping
'''
```

```python
#6. Basic Ansible Playbook

#Ansible uses YAML files called playbooks.

#Create a file named install_nginx.yml

'''
install_nginx.yml
---
- name: Install Nginx
  hosts: localhost
  become: yes

  tasks:
    - name: Install nginx package
      apt:
        name: nginx
        state: present'''
```

```python
#7. Execute Playbook

#Run the playbook using:

'''ansible-playbook -i hosts install_nginx.yml
'''
#This command installs Nginx on the local system.

#8. Another Simple Playbook Example

#Create a file named create_file.yml
'''

create_file.yml
---
- name: Create a file
  hosts: localhost

  tasks:
    - name: Create text file
      file:
        path: /tmp/demo.txt
        state: touch
Run the Playbook
ansible-playbook -i hosts create_file.yml
'''
```

```python
#9. Basic Ansible Commands
'''
ansible --version → Check Ansible version
ansible localhost -m ping → Test connection
ansible-playbook file.yml → Run playbook
ansible-inventory --list → Show inventory
'''
```
