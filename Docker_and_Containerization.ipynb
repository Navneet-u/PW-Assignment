Assignment: Docker and Containerization
1. Introduction
What is Containerization?
Containerization means packing an application with all its files, libraries, and dependencies so it can run on any computer without problems.
The packaged application is called a container.

2. What is Docker?
Docker is a tool used to create, run, and manage containers.
Docker helps developers run applications easily on different systems.

3. Advantages of Docker


Easy to use


Fast application deployment


Works on any system


Lightweight compared to Virtual Machines


Useful in DevOps and Cloud Computing



Practical Part
4. Install Docker
For Windows / Mac


Open the official Docker website:
Docker Desktop Download


Download Docker Desktop.


Install the software.


Restart the system if needed.


Open terminal or command prompt and check Docker version:


docker --version
Example Output:
Docker version 27.0.0

For Ubuntu/Linux
Step 1: Update packages
sudo apt update
Step 2: Install Docker
sudo apt install docker.io -y
Step 3: Start Docker
sudo systemctl start docker
Step 4: Check Docker version
docker --version

5. Create Simple Python Application
Step 1: Create Folder
mkdir docker-appcd docker-app

Step 2: Create Python File
Create a file named app.py
app.py
print("Hello Docker")

6. Create Dockerfile
Create a file named Dockerfile
Dockerfile
FROM python:3.11WORKDIR /appCOPY . .CMD ["python", "app.py"]

7. Build Docker Image
Run this command inside the project folder:
docker build -t my-python-app .
Explanation:


docker build → builds Docker image


-t → gives image name


. → current folder



8. Run Docker Container
docker run my-python-app
Output
Hello Docker
